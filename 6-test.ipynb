```python
import csv
import matplotlib.pyplot as plt
import math

f = open('age-mf.csv', encoding='utf8')
data = csv.reader(f)
header = next(data)

m = []
f = []
size = []

name = ('부산광역시')

for row in data:
    #print(row[3:103])
    #print(row[106:])
    if name in row[0]:
        for i in range(3, 104): #0살 ~ 100살까지
            m.append(int(row[i].replace(',',''))) # 0~100살 남자 추가
            f.append(int(row[i+103].replace(',',''))) # 0~100살 여자 추가
            size.append(math.sqrt(int(row[i].replace(',',''))+int(row[i+103].replace(',','')))) # 총 사이즈 추가
        break

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic') # 한글깨짐 방지
plt.figure(figsize=(10,5), dpi=300)
plt.title(name + " 지역의 성별 인구 그래픽")
plt.scatter(m, f, s = size, c = range(101), cmap = 'jet')
plt.colorbar() # 우측 컬러바
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
