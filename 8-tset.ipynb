```python
import csv
import matplotlib.pyplot as plt
f = open('pop.csv','r',encoding='utf8')
data = csv.reader(f) # 
header = next(data) #   
next(data)

#   
sum_male= []
male = []
m_result = []
sum_female = []
female = []
f_result = []
name = []


# for  
for row in data :
    names = row[0].split(' ')
    name.append(names[0])
    #   SUM        
    #68 104 ( 0~100)
    for i in row[3:104] :
        sum_male.append(int(i))
    male.append(sum(sum_male))
    m_result.append(sum(sum_male))
    
    #( 0~100)
    for j in row[106:202] :
        sum_female.append(int(j))
    female.append(sum(sum_female))
    f_result.append(sum(sum_female))



for m in range(0,17):
    #              
    if male[0] != male[m] and m_result[0] != m_result[m] :
        male[m] -= m_result[m-1]
    if female[0] != female[m] and f_result[0] != f_result[m] :
        female[m] -= f_result[m-1]

#  
plt.figure(figsize=(17,7))
plt.style.use('default')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('   ')
plt.plot(name,female,color='green',marker='^', label='')
plt.plot(name,male,color='red',marker='o',label='')
plt.legend()
plt.show()
plt.figure(figsize=(17,7))
plt.style.use('default')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('   ')
plt.barh(name,female,color='green', label='')
plt.barh(name,male,color='red',label='')
plt.legend()
plt.show()
```
