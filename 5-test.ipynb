```python
import csv
import matplotlib.pyplot as plt
f = open('pop.csv','r',encoding='utf8')
data = csv.reader(f) # 
header = next(data) #   
next(data)

#   
sum_male= []
male = []
m_result = []
sum_female = []
female = []
f_result = []
name = []


# for  
for row in data :
    names = row[0].split(' ')
    name.append(names[0])
    #   SUM        
    #68 104 ( 65~100)
    #      (-)      
    for i in row[68:104] :
        sum_male.append(int(i))
    male.append(sum(sum_male))
    m_result.append(sum(sum_male))
    
    #( 65~100)
    for j in row[171:202] :
        sum_female.append(int(j))
    female.append(sum(sum_female))
    f_result.append(sum(sum_female))
for m in range(0,17):
   
 #              
    if male[0] != male[m] and m_result[0] != m_result[m] :
        male[m] -= m_result[m-1]
    if female[0] != female[m] and f_result[0] != f_result[m] :
        female[m] -= f_result[m-1]

#  
plt.figure(figsize=(17,8))
plt.style.use('default')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(' 65   ')
plt.pie(m_result, labels=name, autopct='%.1f%%') #piechart  
plt.show()
plt.figure(figsize=(17,8))
plt.title(' 65   ') #  piechart   
plt.pie(f_result, labels=name, autopct='%.1f%%')
plt.show()
```
